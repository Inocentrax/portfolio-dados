Sobre o conjunto de dados
📌 Context
O Steam é a maior e mais influente plataforma de distribuição digital de jogos para PC do mundo. Com milhões de jogadores ativos e dezenas de milhares de jogos únicos, que vão desde obras-primas independentes a grandes sucessos AAA, a plataforma gera uma enorme quantidade de dados valiosos.

Este conjunto de dados fornece estatísticas detalhadas extraídas diretamente da API do SteamSpy . Inclui informações sobre desenvolvedores, distribuidoras, bases de jogadores (número estimado de proprietários), métricas de tempo de jogo, número de avaliações (positivas e negativas) e preços de milhares dos jogos mais populares da plataforma.

Se você se interessa por análises de negócios sobre preços de videogames, análise de sentimento sobre a proporção de avaliações positivas e negativas ou pela identificação das "joias escondidas" da plataforma com alto tempo de jogo, mas baixa taxa de aquisição, este conjunto de dados é para você!

# Se subtrairmos positivos por negativos quais itens ainda se posicionariam bem?

```python
df.groupby('nome')['coluna_filtrada'].sum().sort_values(ascending=False)
```
