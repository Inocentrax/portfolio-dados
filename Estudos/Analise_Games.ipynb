#Preparação dos dados


 Fazer uma calculadora que calcula a porcentagem dos dados que tenho em dados nulos e apartir dai saber se posso excluir ou se tem importancia

* motivo: ter certeza do que estou fazendo(baseado em dados concretos)

```python
games.isnull().sum()
```

```python
#vou precisar do total de nulos
nulos_Year = games['Year'].isnull().sum()
nulos_Publisher = games['Publisher'].isnull().sum()
```

```python
#vou precisar do total de linhas
games_rows = games.shape[0]
games_rows
```

```python
#vou querer saber a porcentagem que os dados nulos tem no total.
porcentagem_nulos_Year = nulos_Year/games_rows
porcentagem_nulos_Publisher = nulos_Publisher/games_rows
```

```python
porcentagem_nulos_Year
```

```python
porcentagem_nulos_Publisher
```

* Nem mesmo 1% é de dados nulos, podemos excluir sem medo nulos em Year e Publisher

```python
games.dropna(inplace=True)
```
