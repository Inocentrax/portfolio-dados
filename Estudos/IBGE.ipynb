## Corrigir:

Linha 0, com titulos de colunas.

```python
new_header = df.iloc[0]

df = df[1:]

df.columns = new_header

df = df.reset_index(drop=True)
```

## Corrigir:

Colunas (mês codigo) e coluna (mês) possuem informações duplicadas

1.   criar coluna ano
2.   criar coluna mês
3.   tirar primeiros 4 digitos(ano) - para coluna ano
4.   tirar ultimos 2 digitos(mês) - para coluna mês



```python
df.rename(columns={"Mês (Código)":"ano_mes"}, inplace= True)
```

```python
df["ano_mes"] = df["ano_mes"].astype(str)
```

```python
df["ano"] = df["ano_mes"].str[:4].astype(int)
df["mes"] = df["ano_mes"].str[4:6].astype(int)
```
