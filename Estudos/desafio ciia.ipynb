```python
from unidecode import unidecode

dataframes_tratamento = [df_train, df_test]

for df in dataframes_tratamento:
  for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.lower()
    df[col] = df[col].apply(unidecode)
    df[col] = df[col].str.strip()
    df[col] = df[col].str.replace(r"[^\w\s]", "", regex=True)
```

# Tokenização

Download dos Stopwords da lingua portuguesa-Remoção de Stopwords

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stopwords_en = set(stopwords.words('english'))
df_train['tokens'] = df_train['text'].apply(lambda x: str(x).lower().split())
df_test['tokens'] = df_test['text'].apply(lambda x: str(x).lower().split())

df_train['tokens_sem_stopwords'] = df_train['tokens'].apply(
    lambda tokens: [palavra for palavra in tokens if palavra not in stopwords_en]
)
df_test['tokens_sem_stopwords'] = df_test['tokens'].apply(
    lambda tokens: [palavra for palavra in tokens if palavra not in stopwords_en]
)
```

Exemplo de como a tokenização funcionou

```python
df_train[['text', 'tokens', 'tokens_sem_stopwords']].sample(5)
```

# Vetorização

Usaremos o IDF - Raridade de palavras em outros Tweets - Frequencias Unicas

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Junta os tokens em strings novamente
df_train['texto_final'] = df_train['tokens_sem_stopwords'].apply(lambda x: ' '.join(x))
df_test['texto_final'] = df_test['tokens_sem_stopwords'].apply(lambda x: ' '.join(x))
```

```python
# Cria o vetor de TF-IDF
vetorizador = TfidfVectorizer()

# Ajusta e transforma os textos de treino
X_train_tfidf = vetorizador.fit_transform(df_train['texto_final'])

# Apenas transforma o teste (sem ajustar de novo!)
X_test_tfidf = vetorizador.transform(df_test['texto_final'])
```

# Modelo de Regressão

```python
# Importa a biblioteca do modelo
from sklearn.linear_model import LogisticRegression

# Define os dados de entrada e saída
X = X_train_tfidf
y = df_train['target']

# Cria o modelo
modelo = LogisticRegression(max_iter=1000)

# Treina (ajusta) o modelo com os dados
modelo.fit(X, y)
```

```python
from sklearn.metrics import classification_report, accuracy_score

# Faz previsões com os dados de treino
y_pred = modelo.predict(X)

# Avalia o desempenho
print("Acurácia:", accuracy_score(y, y_pred))
print(classification_report(y, y_pred))
```

```python
# Aplica o modelo treinado no dataset de teste
y_test_pred = modelo.predict(X_test_tfidf)

# Adiciona as previsões no df_test
df_test['target_pred'] = y_test_pred

# Visualiza alguns resultados
df_test[['text', 'target_pred']].sample(5)
```
