* Uso de F´Strings é Essencial para manter a organização

```python
for cada_item in range(len(frutas)):
  print( cada_item, frutas[cada_item])
  print(f'{cada_item}:{frutas[cada_item]}')
```

```python
# Jeito facil sem complicar
for indice, carro in enumerate(carros):
  print(indice, carro)
  print(f'{indice}-{carro}')
```

```python
numeros = [1,30,25,35,78,99,5,18,27] #Lista com numeros impares e pares
pares = [] #lista vazia onde vamos alocar numeros pares
impares = []#lista vazia onde vamos alocar numeros impares
```

```python
for numero in numeros: #Entendendo funcionamento e logica
  if numero % 2 == 0:
    print(f'{numero} é par!')
  else:
    print(f'{numero} é impar!')
```

```python
for numero in numeros:
  if numero % 2 == 0:
    pares.append(numero)

print(pares)
```

```python
for numero in numeros:
  if numero % 2 != 0:
    impares.append(numero)

print(impares)
```

```python
pares1=[]
impares1=[]

for numero in numeros:
  if numero % 2 ==0:
    pares1.append(numero)
  else:
    impares1.append(numero)

print(f'Pares: {pares1}')
print(f'Impares: {impares1}')
```

```python
#vamos utilizar a mesma lista de numeros e iremoms alocar os valores ao quadrado em outra lista vazia, Seguindo a mesma logica
for numero in numeros:
  print(numero ** 2) #Esta é a logica de valores a alocar
```

```python
quadrado = []
for numero in numeros:
  quadrado.append(numero ** 2)
  print(f'Observe a sequencia de iteração que o for vai fazendo {quadrado}')
  print('O for passa numero a numero elevando e alocando com append')
```

```python
print(f'Esta é a visualização da lista após a sequencia de iteração:{quadrado}')
```

* Ordenando a lista - Organizando

Ordenando por ordem alfabetica

```python
linguagens.sort() #ordem alfabetica
linguagens
```

```python
linguagens.sort(reverse=True) #ordem alfabetica inversa
linguagens
```

* Para entender sobre lambda, pesquisar sobre funções anonimas.
argumento x=cada item da lista, e oque queremos é o tamanho de cada item ordenado com ou sem reverse.





ordenando por ordem de tamanho

```python
linguagens.sort(key=lambda x: len(x)) #ordem por tamanho de palavra (crescente)
linguagens
```

```python
linguagens.sort(key=lambda x: len(x), reverse=True) #ordem por tamanho de palavra (decrescente)
linguagens
```

```python
len(linguagens)
```
