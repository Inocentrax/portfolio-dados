# Exercicio 01
DESAFIO DE HOJE — NÍVEL BASE PROFISSIONAL

Você recebe um DataFrame com vendas:

```python
exercicio_1 = pd.DataFrame({
    "produto": ["A", "B", "A", "C", "B", "A", "C", "C"],
    "categoria": ["tech", "tech", "tech", "casa", "tech", "tech", "casa", "casa"],
    "preco": [100, 200, 100, 300, 200, 100, 300, 300],
    "quantidade": [1, 2, 1, 3, 1, 2, 1, 2],
    "cliente_id": [10, 11, 12, 10, 13, 11, 14, 15]
})
```

- Vende mais itens para casa ou tech? total
- Essas vendas vendem em mais quantidades para casa ou tech?
- Quais produtos vendem mais?

Você esqueceu a métrica financeira.

Você olhou:

quantidade

produto

categoria

Qual categoria gera mais receita?

Qual produto gera mais receita?

Qual cliente gera mais receita?

Receita por categoria é concentrada ou distribuída?

## Crie a coluna receita.

```python
exercicio_1['Receita'] = exercicio_1['preco'] * exercicio_1['quantidade']
exercicio_1
```

## Calcule receita total por categoria e ordene do maior pro menor.

```python
exercicio_1['Receita_total'] = exercicio_1.groupby('categoria')['Receita'].transform('sum')
exercicio_1
```

## Calcule quantidade total por categoria.

```python
exercicio_1.groupby('categoria')['quantidade'].sum()
```

## Calcule receita total por produto e pegue o top 2.

```python
receita_total_produtos = exercicio_1.groupby('produto')['Receita_total'].sum().sort_values(ascending=False)
receita_total_produtos[0:2]
```
