# Interpolação de Variaveis

```python
nome = 'Robert'
idade = 26
profissao = 'Analista'
linguagem = 'python'
```

```python
print(f'Olá me chamo {nome} tenho {idade} e trabalho como {profissao} especialmente com linguagem {linguagem}!')
```

```python
print('Olá me chamo {nome} tenho {idade} e trabalho como {profissao} especialmente com linguagem {linguagem}!'.format(nome=nome, idade=idade, profissao=profissao, linguagem=linguagem))
```

* %d para numeros
* %s para strings

```python
print('Olá me chamo %s, tenho %d, e trabalho como %s, especialmente com linguagem %s!' % (nome, idade, profissao, linguagem))
```

```python
print('Olá me chamo {}, tenho {}, e trabalho como {}, especialmente com linguagem {}!'.format(nome, idade, profissao, linguagem))
```

```python
print('Olá me chamo {3}, tenho {2}, e trabalho como {1}, especialmente com linguagem {0}! kkkkk'.format(nome, idade, profissao, linguagem))
```

```python
dados = {'nome':'robert', 'idade':26}
```

```python
print('nome:{nome}, idade:{idade}'.format(**dados))
```

```python
dados2 = {'nome':'robert', 'idade':26, 'saldo':45.435}
```

```python
print('meu nome é {nome}, minha idade é {idade}, Saldo:{saldo:.2f}'.format(**dados2))
```

```python
print(f'Valor de PI:{PI:5.2f}')#5 espaços
```

```python
print(f'Valor de PI:{PI:.3f}')
```

# Strings triplas são a mesma coisa e um pouco mais!


* Uso pratico
1.   Preservação de recuos.
2.   Mensagem completa em uma variavel que já faz referencia a outra variavel.



```python
nome = 'Robert'

mensagem = f'''
  olá meu nome é {nome},
Eu estou aprendendo python.
    Esta mensagem possui varios recuos.
'''
print(mensagem)
```

* Exemplo jeito facil e jeito que pode te complicar no inicio

```python
print('''

****MENU***
    1.DEPOSITAR
2.SACAR
   3.SAIR
***********
Obrigado por usar nosso sistema!

''')
```

* podemos aproveitar os recuos até para centralizar sem precisar de algo mirabolante

```python
menu ='****MENU***\n'
deposito = '1.DEPOSITAR\n'
saque = '2.SACAR\n'
sair = '3.SAIR\n'
print(f'{menu} {deposito} {saque} {sair} ********** \n Obrigado por usar nosso sistema!')
```
