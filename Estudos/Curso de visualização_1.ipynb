```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
trimestres = np.array(['Trimestre 01','Trimestre 02','Trimestre 03','Trimestre 04'])
vendas = np.array([2300,1750,2010,3150])
```

```python
df = pd.DataFrame({'Trimestre':trimestres, 'Vendas':vendas})
df.head()
```

```python
df.plot(x='Trimestre', y='Vendas', kind='line',title='vendas por trimestre', legend=False)
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['Trimestre'], df['Vendas'])
plt.ylabel('Vendas')
plt.xlabel('Trimestre')
plt.grid(True, linestyle='--', alpha=0.2)
plt.title('Vendas por trimestre')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.title('Vendas por Trimestre')
plt.ylabel('Vendas')
plt.xlabel('Trimestre')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.bar(df['Trimestre'], df['Vendas'])
```

```python
plt.figure(figsize=(6,4))
plt.plot(df['Trimestre'], df['Vendas'], marker='o', linestyle='-', color='red')
plt.title('Vendas por trimestre')
plt.ylabel('Vendas')
plt.xlabel('Trimestre')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
sns.lineplot(x='Trimestre', y='Vendas', data=df, marker='o', color='royalblue')
plt.title('Vendas por trimestre')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
