# Criando Função com a mesma metodologia

```python
lista = ["Robert", "Jesus", "Tais", "Pedro", "Jonas", "Elias"]

def lista_minuscula(lista):
  nomes_minusculos = []
  for posicao, nome in enumerate(lista):
    transformacao_nome_minusculo = nome.lower()
    nomes_minusculos.append(transformacao_nome_minusculo) #Adicionando a uma lista
    print(f"{posicao} - {transformacao_nome_minusculo}")
  return nomes_minusculos
lista_formatada = lista_minuscula(lista)
```

```python
def lista_vogais(lista_formatada):
  for posicao, nome in enumerate(lista_formatada):
    vogais = []
    for letra in nome:
      if letra in "aeiou":
        vogais.append(letra)
    print(f"{posicao} - {nome} - Vogais = {vogais}")

lista_vogais(lista_formatada)
```

codigo avançado

```python
cargo_seguranca = 5
cargo_docente = 16
cargo_diretoria = 1

salario_seguranca = 3000
salario_docente = 6000
salario_diretoria = 12500

def total_empregados():
  total = int(cargo_seguranca+cargo_docente+cargo_diretoria)
  print(f"Total de empregados: {total}")
  return total

def maior_menor_salarios():
  todos_salarios = [salario_diretoria,salario_docente,salario_seguranca]
  menor = min(todos_salarios)
  maior = max(todos_salarios)
  media = maior - menor
  print(f"O maior salario é: R${maior},00 e o  menor é: R${menor},00 a média é: R${media},00")

def media_ponderada():
  total = cargo_seguranca+cargo_docente+cargo_diretoria
  media = (salario_diretoria*cargo_diretoria+salario_docente*cargo_docente+salario_seguranca*cargo_seguranca)/total
  print(f"Media ponderada: {int(media)}")

total_empregados()
maior_menor_salarios()
media_ponderada()
```

A quantidade total de empregados;

```python
q_seguranca = 5
s_seguranca = 3000

q_docente = 16
s_docente = 6000

q_diretoria = 1
s_diretoria = 12500

total_empregados = q_seguranca + q_docente + q_diretoria
print(total_empregados)
diferenca_salario = s_diretoria - s_seguranca
print(diferenca_salario)
media = (q_seguranca*s_seguranca + q_docente*s_docente + q_diretoria*s_diretoria) / (total_empregados)
print(media)
```
